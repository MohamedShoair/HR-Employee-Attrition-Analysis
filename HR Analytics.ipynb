```python
#Count Values in RadioNetworkGeneration
value_counts = df['Education'].value_counts()
value_counts
```

```python
# Rename the columns
new_column_names = {'OverTime': 'Overtime', 'MaritalStatus': 'Marital Status', 'JobRole': 'Job Role' , 'EducationField' : 'Education Field' ,
                    'BusinessTravel' : 'Business Travel' , 'TotalWorkingYears' : 'Total Working Years' , 'Education' : 'Education Level',
                    'NumCompaniesWorked' : 'Number of Companies Worked', 'DistanceFromHome' : 'Distance from Home'}
df = df.rename(columns=new_column_names)
```

```python
correlation_matrix = (df.select_dtypes(include=['int64'])).corr()
plt.figure(figsize=(16, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="rocket", linewidths = 0.5)
plt.show()
```
